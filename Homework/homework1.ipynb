```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1.920, 2.080, 0.001)
print(x)
```

```python
p1 = x**9 - 18*x**8 + 144*x**7 - 672*x**6 +2016*x**5 - 4032*x**4 + 5376*x**3 - 4608*x**2 + 2304*x - 512

fig1 = plt.plot(x, p1)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('p(x) = x^9 - 18x^8 +144x^7 - 672x^6 +2016x^5 - 4032x^4 +5376x^3 - 4608x^2 +2304x - 512')
```

```python
p2 = (x - 2)**9

fig2 = plt.plot(x, p2)
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('p(x) = (x - 2)^9')
```
