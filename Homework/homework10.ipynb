```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Question 1

```python
# Just the regular sine function
def f(x): return np.sin(x)

# m = 3, n = 3 from problem 1(a)
def ra(x): 
    p = x - (7/60)*x**3
    q = 1 + (1/20)*x**2
    return p/q

# m = 4, n = 2 from problem 1(b)
def rb(x): 
    p = x - (7/60)*x**3
    q = 1 + (1/20)*x**2
    return p/q

# m = 2, n = 4 from problem 1(c)
def rc(x): 
    p = x
    q = 1 + (1/6)*x**2 + (7/360)*x**4
    return p/q

# The 6th order Maclaurin series of sin(x)
def maclaurin(x):
    return x - x**3/6 + x**5/120


# Creates the evaluation points for the function and approximations
xeval = np.linspace(0, 5, 10000)

# Calculates the error for each of the methods
aError = abs(ra(xeval) - f(xeval))
bError = abs(rb(xeval) - f(xeval))
cError = abs(rc(xeval) - f(xeval))
macError = abs(maclaurin(xeval - f(xeval)))

# Plots the errors on a logy plot
plt.semilogy(xeval, aError,
             marker='o',
             markersize=2,
             label='Pade approximate (m = 3, n = 3)')
plt.semilogy(xeval, bError,
             marker='o',
             markersize=2,
             label='Pade approximate (m = 4, n = 2)')
plt.semilogy(xeval, cError,
             marker='o',
             markersize=2,
             label='Pade approximate (m = 2, n = 4)')
plt.semilogy(xeval, macError,
             marker='o',
             markersize=2,
             label='Maclaurin Series')

plt.title("Error"); plt.xlabel("x"); plt.ylabel("log|error|"); plt.legend(loc='best')
plt.show()
```
